### 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
std=25**0.5
print(f'Рост человека, равный 190 см, отклоняется от математического ожидания на {(190-178)/std} сигмы')
```

Ответ: Рост человека, равный 190 см, отклоняется от математического ожидания на 2.4 сигмы
