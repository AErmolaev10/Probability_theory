### 1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
import numpy as np
import math
a=np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150 ])
n=len(a)
mean=sum(a)/n
deviations = [(x - mean) ** 2 for x in a]
variance = sum(deviations) / n
var = sum(deviations) / (n-1)
std_dev = math.sqrt(variance)
print(f'Среднее арифметическое: {mean}')
print(f'Среднее квадратичное отклонение: {std_dev}')
print(f'Смещённая оценка дисперсии: {variance}')
print(f'Несмещённая оценка дисперсии: {var}')
```

Ответ: Среднее арифметическое: 65.3;
Среднее квадратичное отклонение: 30.823854398825596;
Смещённая оценка дисперсии: 950.11;
Несмещённая оценка дисперсии: 1000.1157894736842.

### 2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
from math import factorial
```

```python
def combinations(n,k):
    return int(factorial(n)/(factorial(k)*factorial(n-k)))
```

```python
# Первый вариант, когда из первой корзины вытаскиваем все небелый, а из второй 3 белых и один не белый
p1=combinations(3,2)/combinations(8,2) * (combinations(5,3)*combinations(7,1)/combinations(12,4))
# Второй вариант, из первой корзины берём один белый мяч, а из второй 2 белых мяча
p2=(combinations(5,1)*combinations(3,1)/combinations(8,2))*(combinations(5,2)*combinations(7,2)/combinations(12,4))
# Третий вариант, из первой корзины берём 2 белых мяча, из второй один белый
p3=(combinations(5,2)/combinations(8,2))*(combinations(5,1)*combinations(7,3)/combinations(12,4))
print(f'вероятность того, что 3 мяча белые: {p1+p2+p3}')
```

Ответ: вероятность того, что 3 мяча белые: 0.3686868686868687

### 3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?
