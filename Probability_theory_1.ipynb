### 1. Из колоды в 52 карты извлекаются случайным образом 4 карты. 

#### a) Найти вероятность того, что все карты – крести.

```python
print(f'Вероятность достать первую крестовую карту 13/52, вторую 12/51, третью 11/50, четвертую 10/49.\n'
      f'Искомая вероятность: {13/52*12/51*11/50*10/49}')
```

Ответ: 0.0026410564225690276

####  б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

```python
import math  
n=math.factorial(52)/(math.factorial(4)*(math.factorial(52 - 4)))
m=math.factorial(48)/(math.factorial(4)*(math.factorial(48 - 4)))
print(f'Общее число доступных комбинаций: {n}')
print(f'Общее число комбинаций в которых не будет туза: {m}')
print(f'Вероятность, что в выбранных картах не будет туза: {m/n}')
print(f'Вероятность, что в выбранных картах будет хотя бы один туз: {1-m/n}')
```

Ответ: 0.2812632745405854
