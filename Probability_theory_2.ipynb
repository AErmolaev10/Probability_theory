### 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

```python
# Применяем формулу Барнулли
import math  
C = (math.factorial(100)/(math.factorial(85)*math.factorial(15)))
p = 0.8**85
q = 0.2**15
print(f'Искомая вероятность: {C*p*q}')

```

Ответ: Искомая вероятность: 0.048061793700746556

### 3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

```python
# Применяем формулу Барнулли
C = (math.factorial(144)/(math.factorial(74)*math.factorial(70)))
p = 0.5**74
q = 0.5**70
print(f'Искомая вероятность: {C*p*q}')
```

Ответ: Искомая вероятность: 0.048061793700746556
